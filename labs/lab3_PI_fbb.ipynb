What you know: The ratio of the area of the circle to the area of the square is Ï€ / 4.

```
def square(l, ax):
  ax.plot([-l, l], [-l, -l], 'k-')
  ax.plot([-l, l], [l, l], 'k-')
  ax.plot([-l, -l], [-l, l], 'k-')
  ax.plot([l, l], [-l, l], 'k-')
  
```

```
def circle(l, ax):
  theta = np.linspace(0, 360, 100)
  theta *= u.degree
  x = l * np.cos(theta.to(u.rad))
  y = l * np.sin(theta.to(u.rad))
  ax.plot(x, y, 'k-')
  #return x, y
  
```

```
fig, ax = pl.subplots(figsize=(5, 5))
square(1, ax=ax)
circle(1, ax=ax)
```

```
numpy.random.seed(125)
```

```
n = 100000 # the larger this is, the better value for pi
```

```
x = np.random.rand(n)*2 - 1
y = np.random.rand(n)*2 - 1
```

```
ax = pl.figure(figsize=(5,5)).add_subplot()
square(1,ax=ax)
circle(1,ax=ax)
plt.plot(x,y,'.')
```

```
def isin(x, y, r=1):
  #given that the radius of circle is 1 and center is 0:
  dist_c = np.sqrt(x**2 + y**2)
  return dist_c < r


```

```
#this sums up all of the true values, and divides by the total number
#of points, then multiples by four because we know the ratio of area 
#of a circle to area of a square (radius 1, length 2) is pi/4
pi = isin(x,y,r=1).sum() / n *4
```
